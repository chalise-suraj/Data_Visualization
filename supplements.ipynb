```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
 
data=pd.read_csv("Supplements_data.csv")
data
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(
    2, 3,
    figsize=(20,10),
    facecolor="#f5f6fa"
)

fig.suptitle(
    "SUPPLEMENTS EVOLUTION IN NEPAL (2022–2026)",
    fontsize=30,
    fontweight="bold",
    color="#0f172a",
    y=0.98
)

plots = [

("WHEY PROTEIN",
 data["Whey Protein (%)"],
 "#1d4ed8","o","+8%","↑"),

("CREATINE",
 data["Creatine (%)"],
 "#f97316","^","+4%","↑"),

("MASS GAINER",
 data["Mass Gainer (%)"],
 "#dc2626","s","-5%","↓"),

("PRE-WORKOUT",
 data["Pre-Workout (%)"],
 "#15803d","D","0%","→"),

("BCAA / EAA",
 data["BCAA/EAA (%)"],
 "#7e22ce","P","-3%","↓"),

("FISH OIL & VITAMINS",
 data["Fish Oil & Vitamins (%)"],
 "#0f766e","X","-4%","↓")

]

for ax,(title,y,color,marker,change,arrow) in zip(axs.flat,plots):

    ax.set_facecolor("white")

    ax.plot(
        data["Year"],
        y,
        color=color,
        linewidth=2.8,
        marker=marker,
        markersize=8
    )

    ax.set_xlim(2021.8,2026.2)
    ax.set_ylim(0,60)

    ax.set_xticks(data["Year"])
    ax.set_yticks(range(0,61,10))

    ax.grid(axis="y",
            linestyle="--",
            alpha=.35)

    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    ax.tick_params(labelsize=11)

    ax.set_ylabel(
        "Market Demand (%)",
        fontsize=13,
        fontweight="bold"
    )

    # ---------- Title ----------
    ax.text(
        0.5,1.08,
        title,
        transform=ax.transAxes,
        ha="center",
        fontsize=18,
        fontweight="bold",
        color=color
    )

    # underline
    ax.plot(
        [0.30,0.70],
        [1.04,1.04],
        transform=ax.transAxes,
        color=color,
        linewidth=3,
        clip_on=False
    )

    # trend
    ax.text(
        0.78,
        1.08,
        f"{arrow} {change}",
        transform=ax.transAxes,
        fontsize=16,
        fontweight="bold",
        color=color
    )

    # last value
    ax.text(
        data["Year"].iloc[-1]+0.12,
        y.iloc[-1],
        f"{y.iloc[-1]}%",
        fontsize=15,
        fontweight="bold",
        color=color,
        va="center"
    )

# legend
handles = [
    plt.Line2D([],[],color="#1d4ed8",marker='o',linewidth=2,label="Whey Protein"),
    plt.Line2D([],[],color="#f97316",marker='^',linewidth=2,label="Creatine"),
    plt.Line2D([],[],color="#dc2626",marker='s',linewidth=2,label="Mass Gainer"),
    plt.Line2D([],[],color="#15803d",marker='D',linewidth=2,label="Pre-Workout"),
    plt.Line2D([],[],color="#7e22ce",marker='P',linewidth=2,label="BCAA / EAA"),
    plt.Line2D([],[],color="#0f766e",marker='X',linewidth=2,label="Fish Oil & Vitamins")
]

fig.legend(
    handles=handles,
    loc="lower center",
    ncol=6,
    fontsize=13,
    frameon=False,
    bbox_to_anchor=(0.5,0.02)
)

plt.tight_layout(rect=[0,0.08,1,0.95])
plt.show()
```
