```python
import matplotlib.pyplot as plt
import numpy as np


labels = [
    "Youth (<24)",
    "Young Adults (25–34)",
    "Midlife (35–44)",
    "Experienced (45–54)",
    "Seniors (55+)"
]

values = [36.1, 53.0, 7.3, 2.3, 1.3]

descriptions = [
    "Over one-third\nparticipation",
    "Largest participation\nshare",
    "Moderate\nparticipation",
    "Lower participation\nshare",
    "Smallest\nparticipation share"
]

colors = [
    "#163C72",
    "#6D9FDC",
    "#4D9AA4",
    "#7A9CC6",
    "#C9DDF2"
]

explode = [0, 0, 0, 0.05, 0.08]


fig = plt.figure(figsize=(14,10), facecolor="#F3F5F7")
ax = plt.subplot(111)
ax.set_facecolor("#F3F5F7")


wedges, texts, autotexts = ax.pie(
    values,
    colors=colors,
    startangle=90,
    counterclock=False,
    explode=explode,
    wedgeprops=dict(width=0.42, edgecolor="white", linewidth=3),
    autopct="%1.2f%%",
    pctdistance=0.78,
    textprops=dict(
        color="white",
        fontsize=18,
        weight="bold"
    )
)


ax.text(
    0, 0.08,
    "89%",
    ha="center",
    va="center",
    fontsize=54,
    fontweight="bold",
    color="#0B1B46"
)

ax.text(
    0, -0.18,
    "are under 35",
    ha="center",
    fontsize=22,
    color="#333"
)


for i, w in enumerate(wedges):

    ang = (w.theta1 + w.theta2)/2
    x = np.cos(np.deg2rad(ang))
    y = np.sin(np.deg2rad(ang))

    start = (x*1.0, y*1.0)
    end = (x*1.18, y*1.18)

    if x >= 0:
        hx = 1.22
        ha = "left"
    else:
        hx = -1.22
        ha = "right"

    # connector
    ax.plot(
        [start[0], end[0], hx],
        [start[1], end[1], end[1]],
        lw=2,
        color="#5D8CC4"
    )

    ax.scatter(
        hx,
        end[1],
        s=110,
        color="#5D8CC4",
        zorder=5
    )

    ax.text(
        hx + (0.07 if x>0 else -0.07),
        end[1]+0.12,
        labels[i],
        fontsize=18,
        weight="bold",
        color="#5D8CC4",
        ha=ha
    )

    ax.text(
        hx + (0.07 if x>0 else -0.07),
        end[1]-0.02,
        f"{values[i]:.2f}%",
        fontsize=28,
        weight="bold",
        color="#0B1B46",
        ha=ha
    )

    ax.text(
        hx + (0.07 if x>0 else -0.07),
        end[1]-0.23,
        descriptions[i],
        fontsize=17,
        color="#333",
        ha=ha
    )


plt.suptitle(
    "Different Age Group Participation",
    fontsize=42,
    fontweight="heavy",
    color="#0B1B46",
    y=0.98
)

plt.title(
    "Two-thirds concentrated in the 25–34 and under–24 bands",
    fontsize=24,
    color="#4B5563",
    pad=30
)

ax.set_aspect("equal")
plt.tight_layout()
plt.show()
```

```python
new_data=pd.read_csv("trainers.csv")
```

```python

years = new_data["Year"]
cert = new_data["Certified Trainers (%)"]
non = new_data["Non-Certified Trainers (%)"]

plt.rcParams["font.family"] = "DejaVu Sans"

fig, ax = plt.subplots(figsize=(16,8), facecolor="#F4F6F8")
ax.set_facecolor("white")


ax.fill_between(
    years,
    cert,
    color="#2563EB",
    alpha=0.18,
    zorder=1
)

ax.fill_between(
    years,
    non,
    color="#DC2626",
    alpha=0.18,
    zorder=1
)

ax.plot(
    years,
    cert,
    color="#2563EB",
    linewidth=3,
    label="Certified Trainers",
    zorder=3
)

ax.plot(
    years,
    non,
    color="#DC2626",
    linewidth=3,
    label="Non-Certified Trainers",
    zorder=3
)


ax.scatter(
    years,
    cert,
    s=120,
    color="white",
    edgecolors="#2563EB",
    linewidth=3,
    zorder=4
)

ax.scatter(
    years,
    non,
    s=120,
    color="white",
    edgecolors="#DC2626",
    linewidth=3,
    zorder=4
)


for x, y in zip(years, cert):
    ax.text(
        x,
        y+3,
        f"{y}",
        color="#2563EB",
        fontsize=15,
        fontweight="bold",
        ha="center"
    )

for x, y in zip(years, non):
    ax.text(
        x,
        y+3,
        f"{y}",
        color="#DC2626",
        fontsize=15,
        fontweight="bold",
        ha="center"
    )


ax.set_ylim(0,100)
ax.set_xticks(years)
ax.tick_params(labelsize=15)

ax.grid(
    axis="both",
    linestyle="--",
    linewidth=.8,
    alpha=.35
)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

fig.text(
    .07,
    .93,
    "ACTIVE TRAINERS DATA",
    fontsize=34,
    fontweight="heavy",
    color="#0F172A"
)

fig.text(
    .07,
    .88,
    "Trends in Certified vs. Non-Certified Trainers (2022–2026)",
    fontsize=18,
    color="#64748B"
)


fig.add_artist(
    plt.Line2D(
        [0.07,0.12],
        [0.845,0.845],
        color="#2563EB",
        linewidth=4,
        solid_capstyle="round"
    )
)


ax.legend(
    loc="lower center",
    bbox_to_anchor=(0.5,-0.18),
    ncol=2,
    frameon=False,
    fontsize=15
)

plt.tight_layout(rect=[0,0,0.75,0.82])
plt.show()
```

```python

years = new_data["Year"]
values = new_data["Yearly Increase in Certified Trainers (%)"]

highest = values.max()
average = values.mean()


fig = plt.figure(figsize=(16,9), facecolor="#F3F5F9")

plt.figtext(
    0.05,0.94,
    "Growth of Certified Trainers (2022–2026)",
    fontsize=34,
    fontweight="bold",
    color="#091B4A"
)

plt.figtext(
    0.05,0.88,
    "Percentage of certified trainers across five years.",
    fontsize=18,
    color="#5F6368"
)

plt.figtext(
    0.05,0.865,
    "━━━━",
    fontsize=28,
    color="#1565F5"
)


ax1 = fig.add_axes([0.05,0.70,0.43,0.12])
ax2 = fig.add_axes([0.52,0.70,0.43,0.12])

for ax in [ax1,ax2]:
    ax.set_facecolor("white")
    ax.set_xticks([])
    ax.set_yticks([])
    for s in ax.spines.values():
        s.set_color("#E3E3E3")

ax1.text(.15,.52,f"{highest:.1f}%",
         fontsize=34,
         color="#1565F5",
         weight="bold",
         transform=ax1.transAxes)

ax1.text(.50,.52,"Highest Achieved",
         fontsize=20,
         transform=ax1.transAxes)

ax1.plot([.45,.45],[.25,.75],
         color="#BBBBBB",
         transform=ax1.transAxes)

ax2.text(.15,.52,f"{average:.1f}%",
         fontsize=34,
         color="#1565F5",
         weight="bold",
         transform=ax2.transAxes)

ax2.text(.50,.52,"5-Year Average",
         fontsize=20,
         transform=ax2.transAxes)

ax2.plot([.45,.45],[.25,.75],
         color="#BBBBBB",
         transform=ax2.transAxes)

ax = fig.add_axes([0.08,0.10,0.84,0.46])

ax.set_facecolor("white")


ax.fill_between(
    years,
    values,
    color="#1565F5",
    alpha=.15
)


ax.plot(
    years,
    values,
    color="#1565F5",
    linewidth=3
)

ax.scatter(
    years,
    values,
    s=260,
    color="#1565F5",
    edgecolor="white",
    linewidth=2.5,
    zorder=5
)


for x,y in zip(years,values):
    ax.vlines(
        x,
        10,
        y,
        linestyle="--",
        color="#BFBFBF",
        linewidth=1
    )

# labels
for x,y in zip(years,values):
    ax.text(
        x,
        y+0.9,
        f"{y:.1f}%",
        fontsize=16,
        fontweight="bold",
        color="#1565F5",
        ha="center"
    )

# formatting
ax.set_xlim(2021.7,2026.3)
ax.set_ylim(10,30)

ax.set_xticks(years)
ax.set_yticks(np.arange(10,31,2))

ax.tick_params(labelsize=16)

ax.grid(
    axis="y",
    linestyle="--",
    alpha=.35
)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_ylabel(
    "Percentage (%)",
    fontsize=18,
    fontweight="bold"
)

plt.show()
```
